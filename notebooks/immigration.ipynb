```python
import os
from os.path import join, dirname
import sys

SOURCE_CODE_PARENT = join(
    dirname(os.getcwd()),
    'src'
)

DATA_PARENT = join(
    dirname(os.getcwd()),
    'data'
)

sys.path.append(SOURCE_CODE_PARENT)
```

```python
import pandas as pd

csv_name = 'immigration_data.csv'
df = pd.read_csv(
    join(DATA_PARENT, csv_name)
)
df.head()
```

```python
df['date'] = df['created_at'].apply(lambda x: x[:10])
df.head()
```

```python
from alerts import preprocess

freq_per_day_df = preprocess.data2timeSeries(df)
freq_per_day_df.head()
```

```python
df.index.values[:4], df.text.values[:4]
```

```python
from alerts.time_series import AnomalyTS

T = zip(freq_per_day_df.index.values, freq_per_day_df.text.values)
parameters = {
    'window': 30,
    'static_metric': 'mean',
    'transf': None
}
ts = AnomalyTS(T, parameters)
ts.plot(show_anomalies=True)
```

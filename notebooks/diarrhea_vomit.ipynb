```python
import os
from os.path import join, dirname
import sys

SOURCE_CODE_PARENT = join(
    dirname(os.getcwd()),
    'src'
)

DATA_PARENT = join(
    dirname(os.getcwd()),
    'data'
)

sys.path.append(SOURCE_CODE_PARENT)
```

```python
import pandas as pd

csv_name = 'diarrhea_vomit_data.csv'
df = pd.read_csv(
    join(DATA_PARENT, csv_name)
)
df.head()
```

```python
df['date'] = df['created_at'].apply(lambda x: x[:10])
df.head()
```

```python
from alerts import preprocess

freq_per_day_df = preprocess.data2timeSeries(df)
freq_per_day_df.head()
```

```python
df.index.values[:4], df.text.values[:4]
```

```python
from alerts.time_series import AnomalyTS

T = zip(freq_per_day_df.index.values, freq_per_day_df.text.values)
parameters = {
    'window': 5
}
ts = AnomalyTS(T, parameters)
ts.plot()
```

```python
T = zip(freq_per_day_df.index.values, freq_per_day_df.text.values)
parameters = {
    'window': 5
}
ts = AnomalyTS(T, parameters)
T = ts.normalize_time_series()
parameters = {
    'window': 5
}
ts = AnomalyTS(T, parameters)
ts.plot()
```

```python
import pandas as pd

T = zip(freq_per_day_df.index.values, freq_per_day_df.text.values)
parameters = {
    'window': 30
}
ts = AnomalyTS(T, parameters)

predictions = ts.prediction_interval()
pred_df = pd.DataFrame(predictions).tail(60)
pred_df.head(50)
```

```python
import matplotlib.pyplot as plt

plt.subplots(figsize=(12, 5))


simplified_preds = pred_df[~pred_df['lb'].isna()]

plt.plot(
    simplified_preds['dates'],
    simplified_preds['values'],
    label='value'
)

plt.plot(
    simplified_preds['dates'],
    simplified_preds['lb'],
    label='lb'
)

plt.plot(
    simplified_preds['dates'],
    simplified_preds['ub'],
    label='ub'
)

plt.xlabel("Angle")
plt.ylabel("Magnitude")
plt.title("Time series")

plt.legend()


plt.show()
```

```python
import matplotlib.pyplot as plt

plt.subplots(figsize=(12, 5))


simplified_preds = pred_df[~pred_df['lb'].isna()]

plt.plot(
    simplified_preds['dates'],
    simplified_preds['flatten_values'],
    label='value'
)

plt.plot(
    simplified_preds['dates'],
    simplified_preds['flatten_lb'],
    label='lb'
)

plt.plot(
    simplified_preds['dates'],
    simplified_preds['flatten_ub'],
    label='ub'
)

plt.xlabel("Angle")
plt.ylabel("Magnitude")
plt.title("Time series (sqrt)")

plt.legend()


plt.show()
```

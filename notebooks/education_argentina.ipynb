```python
import os
from os.path import join, dirname
import sys

SOURCE_CODE_PARENT = join(
    dirname(os.getcwd()),
    'src'
)

DATA_PARENT = join(
    dirname(os.getcwd()),
    'data'
)

sys.path.append(SOURCE_CODE_PARENT)
```

```python
import pandas as pd

csv_name = 'Education_Argentina.csv'
df = pd.read_csv(
    join(DATA_PARENT, csv_name),
    delimiter=';'
)
df.head()
```

```python
import pandas as pd

df['date'] = df['Date'].apply(lambda x: x[:10])
df['date'] = pd.to_datetime(df.loc[:, "date"])
df['total'] = df['Total docs']
df.head()
```

```python
from alerts.time_series import AnomalyTS

T = zip(df.date, df.total)
parameters = {
    'window': 30,
    'static_metric': 'mean',
    'transf': None
}
ts = AnomalyTS(T, parameters)
ts.plot(show_anomalies=True)
```
